```python
arr=np.linspace(-np.pi,np.pi,num=20)
```

```python
plt.plot(np.sin(arr),arr)
plt.plot(np.cos(arr),arr)
plt.show()
```

```python
plt.plot(arr,np.sin(arr))
plt.plot(arr,np.cos(arr))
plt.show()
```

```python
plt.figure(edgecolor='white')
plt.plot(np.sin(arr),arr)
plt.plot(np.cos(arr),arr)
plt.show()
```

```python
plt.plot(arr,np.sin(arr),color='c',linewidth=2,)
plt.plot(arr,np.cos(arr),color='b',linewidth=2)
plt.show()
```

```python
y_sin=np.sin(arr)
y_cos=np.cos(arr)
plt.plot(arr,y_sin,linestyle='--',marker='o',linewidth=
        3,label='sin(x)')
plt.plot(arr,y_cos,linestyle='-',marker='x',linewidth=
        3,label='cos(x)')
plt.xlim((arr.min())*1.1,(arr.max())*1.1)
plt.ylim(y_sin.min()*1.1,y_sin.max()*1.1)
plt.xticks([-np.pi,-np.pi/2,0,np.pi/2,np.pi],[r'$-\pi$',r'$-\pi/2$',r'$0$',r'$\pi/2$',r'$\pi$'])
plt.yticks([-1,0,1])
plt.legend(loc=0,frameon=False,fontsize='large')
plt.show()
```
